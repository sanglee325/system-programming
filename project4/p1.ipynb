```
def increase(N, A, result):
  max = 0
  for idx in A:
    if int(idx) == N+1:
      for i in range(N):
        if max < result[i]:
          max = result[i]
      for i in range(N):
        result[i] = max
    else:
      result[int(idx)-1] += 1
      
print("Input:")
N = int(input())
result = []
for i in range (N):
  result.append(0)

tmp = input()
A = tmp.split()
intA = []
for ch in A:
  intA.append(int(ch))
increase(N, intA, result)

print("\nOutput:")
for i in result:
  print(i, end=" ")
```

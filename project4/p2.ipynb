```
def get_prime(num, prime_list):
  if num > 1:
    for i in range(2, num+1):
      flag_prime = 1
      for j in range(2, i):
        if (i % j) == 0:
          flag_prime = 0
          break
      if flag_prime == 1:
        prime_list.append(i)
        

def common_prime_div(a, b):
  primeA = []
  primeB = []
  get_prime(a, primeA)
  get_prime(b, primeB)
  
  A_set = set()
  B_set = set()
  for prime in primeA:
    if(a % prime) == 0:
      A_set.add(prime)
  for prime in primeB:
    if(b % prime) == 0:
      B_set.add(prime)   
      
  if A_set == B_set:
    return 1
  else:
    return 0


tmpA = []
tmpB = []
A = []
B = []

print("Input:")
inputA = input()
tmpA = inputA.split()
for a in tmpA:
  A.append(int(a))

inputB = input()
tmpB = inputB.split()
for b in tmpB:
  B.append(int(b))

result = 0
for idx in range(len(A)):
  result += common_prime_div(A[idx], B[idx])
print("\nOutput:")
print(result)
```

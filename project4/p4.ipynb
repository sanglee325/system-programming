```
import numpy as np

def maxpool(n, A):
  result = np.zeros((int(n/2), int(n/2)))
  
  for idx1 in range(int(n/2)):
    i = idx1*2
    for idx2 in range(int(n/2)):
      j = idx2*2
      result[idx1, idx2] = np.max(A[i:i+2, j:j+2])
    
  return result
    

print("Input:")
n = int(input())

A = np.random.random((n, n))
print("\nOutput:")
print(A)

B = maxpool(n, A)
print(B)
```
